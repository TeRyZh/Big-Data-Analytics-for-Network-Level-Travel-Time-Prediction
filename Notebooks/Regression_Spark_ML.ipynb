```python
travel_time_df1 = pd.read_csv('/content/drive/MyDrive/CIS 545 Project Folder/Data Set/2021_travel_time_15min.csv/part-0.csv')
travel_time_df2 = pd.read_csv('/content/drive/MyDrive/CIS 545 Project Folder/Data Set/2021_travel_time_15min.csv/part-1.csv')
travel_time_df = pd.concat([travel_time_df1, travel_time_df2])

corridor_ids = set(list(travel_time_df1.corridor_id.unique())  + list(travel_time_df2.corridor_id.unique()))
corridor_ids = list (corridor_ids)

travel_time_df
```

```python
travel_time_df['datetime'] = travel_time_df['datetime'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
# split datetime into month, day of week, and hour
travel_time_df ['Month'] = pd.DatetimeIndex(travel_time_df['datetime']).month
travel_time_df ['Weekday'] = pd.DatetimeIndex(travel_time_df['datetime']).dayofweek
travel_time_df ['Hour'] = pd.DatetimeIndex(travel_time_df['datetime']).hour
```

```python
# Convert the month, day of week, and hour features in the dataset into one hot vectors
travel_time_df.Month = travel_time_df.Month.astype('category')
travel_time_df = pd.get_dummies(travel_time_df,columns = ['Month'])
travel_time_df.Weekday= travel_time_df.Weekday.astype('category')
travel_time_df = pd.get_dummies(travel_time_df,columns = ['Weekday'])
travel_time_df.Hour = travel_time_df.Hour.astype('category')
travel_time_df = pd.get_dummies(travel_time_df, columns = ['Hour'])
```

```python
train_sdf = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/content/drive/MyDrive/CIS 545 Project Folder/Data Set/2021_tt_train_15min.csv")
```

```python
train_sdf.show(5)
```

```python
all_columns = train_sdf.columns
drop_columns = ['datetime', 'next_tt', "_c0", "corridor_id"] 
columns_to_use = [i for i in all_columns if i not in drop_columns]
assembler = VectorAssembler(inputCols= columns_to_use, outputCol="features")
```

```python
train_sdf.createOrReplaceTempView("train_sdf")
```

###ElasticNet 15 Minute

```python
rmse_eval = RegressionEvaluator(labelCol="next_tt", predictionCol="prediction",metricName="rmse")
mae_eval = RegressionEvaluator(labelCol="next_tt", predictionCol="prediction",metricName="mae")
r2_eval = RegressionEvaluator(labelCol="next_tt", predictionCol="prediction",metricName="r2")

final_evaluations = []

for corridor_id in corridor_ids:

    sub_df = spark.sql(f"SELECT * FROM train_sdf where corridor_id = {corridor_id}")
    sub_df.createOrReplaceTempView("sub_df")
    data_sdf = assembler.transform(sub_df)
    train_sdf, test_sdf = data_sdf.randomSplit([0.8,0.2])

    elastic_net_predictions = LinearRegression(featuresCol = 'features', labelCol = 'next_tt',regParam=0.3, elasticNetParam=0.8).fit(train_sdf).transform(test_sdf)
    test_rmse_elastic = rmse_eval.evaluate(elastic_net_predictions)
    test_mae_elastic = mae_eval.evaluate(elastic_net_predictions)
    test_r2_elastic = r2_eval.evaluate(elastic_net_predictions)
    pandasDF = elastic_net_predictions.toPandas()
    test_mape_elastic = mean_absolute_percentage_error(pandasDF["next_tt"].to_numpy() , pandasDF["prediction"].to_numpy())
    
    final_evaluations.append([corridor_id, test_r2_elastic, test_rmse_elastic, test_mae_elastic, test_mape_elastic])

df = pd.DataFrame(final_evaluations, columns =['corridor_id', 'r2', 'rmse', 'mae', 'mape'])
df.to_csv("/content/drive/MyDrive/CIS 545 Project Folder/Data Set/ElasticNet_records_15min.csv")
df[['r2', 'rmse', 'mae', 'mape']].mean(axis=0)
```
